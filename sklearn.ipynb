### 1. 일반적인 분류 (DecisionFree활용)  
- 데이터 준비하기 
- 의사결정트리 클래스를 객체로 생성
- train/ test 데이터 분리
- 의사결정트리에 학습용 데이터,레이블 학습
- test용 데이터 사용하여 예측
- 평가

```python
from sklearn.metrics import accuracy_score

#데이터 준비
iris=load_iris()

#의사결정트리 클래스를 객체(clf)로 생성
clf=DecisionTreeClassifier()

#train/ test 데이터 분리
X_train,X_test,y_train,y_test=train_test_split(
    iris.data, iris.target, test_size=0.3, random_state=121)

#의사결정트리에 학습용 피처,레이블 학습
clf.fit(X_train, y_train)

#test용 데이터 예측
pred=clf.predict(X_test)

#평가: 성능평가지표 
print('성능평가지표 :',accuracy_score(y_test,pred))
print('성능평가지표 : {0:.4f}'.format(accuracy_score(y_test,pred)))
```

### 2. 분류 + 교차검증  
    1) KFord  
    2) StratifiedKFold 
    3) cross_val_score()
- 

    1) KFold : 피처, 레이블이 인덱스로 반환됨 (train index / test index)

```python
from sklearn.model_selection import KFold
import numpy as np

#iris=load_iris()
features=iris.data
label=iris.target
clf=DecisionTreeClassifier(random_state=156)

#kfold객체와 세트별 성능평가(정확도)를 담을 리스트 생성
kfold=KFold(n_splits=5)
cv_accuracy=[]

#for문 사용해서 폴드별(인덱스)로 학습/검증 실행
for train_index, test_index in kfold.split(features):

    #실행별 test_index 확인
    print(f'test_index : {test_index[[0,-1],]}')
    
    #split으로 반환된 인덱스를 활용해 학습/검증용 테스트 데이터 추출
    X_train,X_test=features[train_index], features[test_index]
    y_train,y_test=label[train_index], label[test_index]
    
    #학습
    clf.fit(X_train, y_train)
    #예측
    pred=clf.predict(X_test)
    
    #반복 마다 정확도 측정
    accuracy=np.round(accuracy_score(y_test,pred),4)
    
    #폴드별 정확도를 리스트에 추가
    cv_accuracy.append(accuracy)
    
print(f'폴드 별 정확도 : {cv_accuracy}\n평균 정확도 : {np.mean(cv_accuracy)}')
```

    2) StratifiedKFold : test데이터의 레이블 비율(산포도)을 전체 데이터의 비율과 동일하게 구성
            for문 사용시 features, label 모두 사용해야함

```python
from sklearn.model_selection import StratifiedKFold

#iris=load_iris()
# features=iris.data
# label=iris.target
#clf=DecisionTreeClassifier(random_state=156)

#StratifiedKFold객체와 세트별 성능평가(정확도)를 담을 리스트 생성
skf=StratifiedKFold(n_splits=5)
cv_accuracy2=[]

for train_idx, test_idx in skf.split(features,label):
    
    print(f'test_index : {test_idx}')

    #split으로 반환된 인덱스를 활용해 학습/검증용 테스트 데이터 추출
    X_train, X_test= features[train_idx],features[test_idx]
    y_train, y_test= label[train_idx],label[test_idx]
    
    #학습
    clf.fit(X_train, y_train)
    
    #평가
    pred=clf.predict(X_test)
    
    #폴드마다 정확도 측정
    accuracy2= np.round(accuracy_score(y_test,pred),4)
    
    #리스트에 정확도 저장
    cv_accuracy2.append(accuracy2)

    
print(f'폴드 별 정확도 : {cv_accuracy2}\n평균 정확도 : {np.mean(cv_accuracy2)}')
```

    3) cross_val_score() : 폴드세트구분, for문 반복(인덱스추출), 학습-예측-평가지표 과정을 한번에 진행
        estimator : classifier 입력시 stratifiedKFold 실행
                    regressor 입력시 KFold 실행

```python
from sklearn.model_selection import cross_val_score, cross_validate


#iris=load_iris()
# features=iris.data
# label=iris.target
#clf=DecisionTreeClassifier(random_state=156)

#cross_val_score 객체 생성 (estimator, X, y, scoring, cv)
        # estimator: 분류/회귀 알고리즘 , X:피처 , y:레이블 , scoring:성능평가지표 , cv= 교차검증
score=cross_val_score(clf, features, label, scoring='accuracy',cv=5)

#교차검증 별 정확도
print(score)

#평균 정확도
print(np.round(np.mean(score),4))
```
